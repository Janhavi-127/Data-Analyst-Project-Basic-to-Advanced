✅ 1. Import necessary libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

✅ 2. Load the dataset

```python
df = pd.read_csv('Global_AI_Content_Impact_Dataset.csv')
```

✅ 3. Preview the data

```python
# View first 5 rows
print(df.head())

# View info about data types & null values
print(df.info())

# Basic statistics
print(df.describe())
```

✅ 4. Check for missing values

```python
print(df.isnull().sum())
```

✅ 5. Analyze categorical columns (if any)

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

✅ 6. Correlation matrix (for numerical columns)

```python
# Select only numerical features for correlation analysis
numerical_features = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

✅ 7. Univariate analysis (distribution of individual variables)

```python
# Example for numeric column
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()

for col in numeric_columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

✅ 8. Bivariate analysis (relation between two variables)

```python
# Example: scatter plot between two numeric variables
sns.scatterplot(data=df, x='Country', y='Industry')
plt.title('Country vs Industry')
plt.show()
```

✅ 9. Grouped analysis (if applicable)

```python
df.groupby('Industry')['AI Adoption Rate (%)'].mean()
```

```python
df.groupby('Industry')['AI Adoption Rate (%)'].mean().sort_values(ascending=False)
```

```python
df.groupby('Industry')['AI Adoption Rate (%)'].mean().sort_values().plot(kind='barh')
plt.title('Average AI Adoption Rate by Industry')
plt.xlabel('AI Adoption Rate (%)')
plt.show()
```

```python
df.groupby('Regulation Status')['Consumer Trust in AI (%)'].mean()
```

✅ 10. Export cleaned data (optional)

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
